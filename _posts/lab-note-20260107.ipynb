## Goal

- Paper0: Böhne, Astrid, et al. "The vertebrate makorin ubiquitin ligase gene family has been shaped by large-scale duplication and retroposition from an ancestral gonad-specific, maternal-effect gene." BMC genomics 11.1 (2010): 721.[https://link.springer.com/article/10.1186/1471-2164-11-721]
  **Species:** Xiphophorus maculatus, medaka, zebrafish and amphibians
  **Mechanism:**  Via partial sequencing of a testis cDNA library from the poeciliid fish Xiphophorus maculatus, we have identified a new member of the makorin gene family, that we called mkrn4. In addition to the already described mkrn1 and mkrn2, mkrn4 is the third example of a makorin gene present in both tetrapods and ray-finned fish. However, this gene was not detected in mouse and rat, suggesting its loss in the lineage leading to rodent murids. Mkrn2 and mkrn4 are located in large ancient duplicated regions in tetrapod and fish genomes, suggesting the possible involvement of ancestral vertebrate-specific genome duplication in the formation of these genes. Intriguingly, many mkrn1 and mkrn2 intronless retrocopies have been detected in mammals but not in other vertebrates, most of them corresponding to pseudogenes. The nature and number of zinc fingers were found to be conserved in Mkrn1 and Mkrn2 but much more variable in Mkrn4, with lineage-specific differences. RT-qPCR analysis demonstrated a highly gonad-biased expression pattern for makorin genes in medaka and zebrafish (ray-finned fishes) and amphibians, but a strong relaxation of this specificity in birds and mammals. All three mkrn genes were maternally expressed before zygotic genome activation in both medaka and zebrafish early embryos.
  **Results:** gonad-specific, maternal-effect

## Mkrn1
- Paper1: Dold, Annabelle, et al. "Makorin 1 controls embryonic patterning by alleviating Bruno1-mediated repression of oskar translation." PLoS genetics 16.1 (2020): e1008581. [https://journals.plos.org/plosgenetics/article?id=10.1371/journal.pgen.1008581]  
  **Species:** Drosophila  
  **Mechanism:** Mkrn1 interacts with poly(A) binding protein (pAbp) and binds specifically to osk 3’ UTR in a region adjacent to A-rich sequences. Mkrn1 controls embryonic patterning and germ cell formation by specifically activating osk translation, most likely by competing with Bru1 to bind to osk 3’ UTR.  
  **Results:** axis specification, pole plasm assembly 

- Paper2: Gray, Todd A., et al. "The putatively functional Mkrn1-p1 pseudogene is neither expressed nor imprinted, nor does it regulate its source gene in trans." Proceedings of the National Academy of Sciences 103.32 (2006): 12039-12044.[https://www.pnas.org/doi/abs/10.1073/pnas.0602216103]
  **Species:** Mice  
  **Mechanism:** The pseudogene of Mkrn1, called Mkrn-p1, regulates stability of the Mkrn1 mRNA in trans. Transgenic mice with reduced transcription of Mkrn1-p1 manifested polycystic kidneys and bone deformity. However, it was shown later that Mkrn1-p1 is fully methylated on both alleles; it is not transcribed and, therefore, cannot stabilize Mkrn1 transcripts in trans. Moreover, none of the phenotypes caused by a partial reduction of Mkrn1 in mutant mice were seen in the mutants with direct Mkrn1 disruption. 
  **Results:** no phenotypes of polycystic kidneys and bone deformity


## Mkrn2
- Paper1: Wang, Lin, et al. "MKRN2 knockout causes male infertility through decreasing STAT1, SIX4, and TNC expression." Frontiers in Endocrinology 14 (2023): 1138096 [https://www.frontiersin.org/journals/endocrinology/articles/10.3389/fendo.2023.1138096/full]  
  **Species:** Mice  
  **Mechanism:** MKRN2 regulated the expression level of SIX4 and tenascin C (TNC) via the EBF transcription factor 2 (EBF2) in mice.  
  **Results:** male infertility, abnormal function of gonad

- Paper2: Qian, Xu, et al. "Deficiency of Mkrn2 causes abnormal spermiogenesis and spermiation, and impairs male fertility." Scientific reports 6.1 (2016): 39318.[https://www.nature.com/articles/srep39318]  
  **Species:** Mice  
  **Mechanism:** Mkrn2 knockout mice produced abnormal sperms characterized by low number, poor motility, and aberrant morphology. To understand the molecular mechanism, we found that expression of Odf2, a vital protein in spermatogenesis, was significantly decreased. In addition, we found that expression levels of Odf2 were decreased in Mkrn2 knockout mice.
  **Results:** male infertility, abnormal spermiogenesis
